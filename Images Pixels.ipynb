```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

### Reading the image

```python
url = "./1.Data/Train Buildings/10006.jpg"
pic =  pltimg.imread(url,0)
print ("Image dimensions:", pic.shape)
```

```python
gray_scale_pic = cv2.cvtColor(pic, cv2.COLOR_RGB2GRAY) #changing from red, blue and green to grayscale
```

```python
plt.imshow(gray_scale_pic, cmap = 'gray')
```
